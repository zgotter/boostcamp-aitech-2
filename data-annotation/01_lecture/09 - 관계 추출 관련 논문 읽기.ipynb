### 9.2.2 Overview

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<img src='https://drive.google.com/uc?id=11ypFximVnhL1yYP4xjm7ok9eBuDVuetf' width=900/>

- RE task에 대한 정리
- KLUE RE는 single sentence classification task로 본다.
- Subject Entity, Object Entity, Relation triplet을 만드는 것이 목적이다.
- relation의 종류 (20개)
  - 18개의 person-related relations
  - 11개의 organization-relation relations
  - 1개의 no_relation
- 모델 평가
  - micro F1 score를 사용 (no_relation은 제외)
  - no_relation 포함 시 AUPRC 사용

<br>

### 9.2.3 Data Construction

#### 9.2.3.1 Collect Candidate Sentences

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<img src='https://drive.google.com/uc?id=12e598Wi8UW_qBc0vi7sC_fIrQes3D0rs' width=900/>

- 주석의 대상이 되는 후보 문장들을 어디서  가져와야 하는 지 명시
- 3가지 원시 데이터 source를 밝히고 있다.
  - WIKIPEDIA
    - 온라인 백과사전
    - 문어적인 성격을 띄고 있음
  - WIKITREE
    - 뉴스 기사
  - POLICY
    - 정부 정책 브리핑 뉴스 기사
- 구어가 아닌 문어성이 짙은 말뭉치들을 사용했다.
- 문장 처리를 할 때는 KSS(Korean Sentence Spliter) 라이브러리를 사용했다.
- Korean hate speech dataset을 통해 hate speech와 관련된 표현을 제거했다.

#### 9.2.3.4 Select Entity Pairs

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<img src='https://drive.google.com/uc?id=12ns51mcSYbfbljloVL1USJ2SHuO_L2li' width=900/>

- 정해진 두 Entity의 관계를 만들 때 2가지 approach를 적용했다.
1. KB-based sampling
  - KB로 된 샘플링 방법
  - WIKIPEDIA와 WIKITREE 데이터셋에 해당 방법 적용
2. uniform sampling
  - real world 단어를 반영하기 위해 no_relation 데이터를 추가하기 위해 임의적인 데이터 생성
