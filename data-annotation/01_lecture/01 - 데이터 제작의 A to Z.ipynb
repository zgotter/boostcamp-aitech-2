# 1. 데이터 제작의 A to Z

## 강의 소개

- 전체 강의 내용을 요약합니다.
- 데이터 구축 프로세스와 설계의 기초 개념에 대해 설명합니다.
    - 데이터 구축이 전체 인공지능 서비스 개발에서 차지하는 역할에 대해 알아봅니다.
    - 데이터 유형, 데이터 입출력 형식, 데이터 구분 방식에 대해 배웁니다.
- 자연어처리 데이터의 특징 및 종류에 대해 설명합니다.

## Further Questions

- 자연어 데이터와 다른 데이터를 구분짓는 특징은 무엇일까요?

### 1.5.2 자연어처리(NLP, Natural Language Processing)란?

- 인공지능의 한분야
- 사람의 언어를 컴퓨터가 알아듣도록 처리하는 인터페이스 역할
- 다음 2가지로 구성되어 있음
  - 자연어 이해 (NLU, Natural Language Understanding)
  - 자연어 생성 (NLG, Natural Language Generation)
- 자연어 처리의 최종 목표
  - 컴퓨터가 사람의 언어를 이해하고 여러 가지 문제를 수행할 수 있도록 하는 것

<br>

### 1.5.3 자연어처리와 관련 연구 분야

- Computer Science(컴퓨터 과학)와 Linguistics(언어학)가 큰 틀로서 자리잡고 있다.
  - 95%의 NLP 연구자는 Computer Science에서 출발한다.
- NLP는 인공지능(Artificial Intelligence)의 한 분야이다.
- NLP를 하기 위해 수학 및 통계적 근거를 활용한다.
- Data Mining 및 Text Mining을 진행하여 데이터 자체에 집중하는 경우도 있다.

### 1.5.4 데이터 분류 방식

- 원천 데이터 장르 (도메인)
  - 문어 (기사, 도서 등)
  - 구어 (대화 등)
  - 웹 (메신저 대화, 게시판 등)
- 과제의 유형
  - 자연어 이해 (형태 분석, 구문 분석, 문장 유사도 평가 등)
  - 자연어 생성 (기계 번역, 추상 요약 등)
  - 혼합 (챗봇 등)
- 자연어처리 데이터를 만들 때는 복잡한 과제도 단순화하여 단계별로 구축
