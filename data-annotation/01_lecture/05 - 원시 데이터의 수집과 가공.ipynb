## 5.1 원시 데이터

### 5.1.1 원시 데이터란?

- 과제를 해결하기 위해 특정 도메인, 장르, 주제 등에 대하여 조건에 맞춰 수집하였으나, 주석 단계를 거치지 않은 상태의 데이터
- 원하는 형태로 가공하기 이전의 데이터로 목적에 맞는 전처리 과정을 거쳐 가공이 되어야 활용할 수 있음

### 5.1.4 원시 데이터 수집 절차 (예시)

- 한국어 대화 텍스트를 크라우드 소싱을 통해 수집한 경우 (왼쪽)
  - 크라우드 워커 모집 단계에서 개인 정보 및 저작권 활용 동의를 얻는 게 중요하다.
  - 대화 요약 데이터 지침 = 가이드 라인
  - 구글 폼 및 구글 스프레드 시트 활용
  - 대화에 참여한 사람들에 대한 정보 제공 동의도 필요
  - 과적합 방지를 위해 작업 할당량 상한선 지정
  - 편향을 제어하는 것도 중요하다.
- 문서 요약 텍스트를 외부 데이터를 통해 수집한 경우 (오른쪽)
  - 이미 존재하는 데이터 수집
  - 기사 데이터는 기사 주제 및 작성 시간과 같은 정보들을 어떻게 처리할 지 정해야 한다.
  - 기고문은 기사와 성격이 다르기 때문에 구분해야 한다.

### 5.1.5 원시 텍스트 데이터의 종류

- 원시 텍스트 데이터 사용역(장르)에 따른 분류

<br>

1. 문어
  - 신문기사, 소설, 수필, 논문, 잡지, 보고서 등
2. 구어
  - 음성 파일을 텍스트로 전사 (굉장히 힘들다...)
  - 일상 대화, 연설, 강연
  - 준구어: 방송 대본, 영화 대본 등 (좀 더 편하게 구어 데이터로서 사용)
3. 웹
  - SNS, 커뮤니티 게시판, 메신저 대화, 블로그, 이메일 등

## 5.2 원시 데이터 수집 시 고려 사항

### 5.2.1 획득 가능성

- 획득이 불가능하거나 통제 불가능한 주기를 가지고 있다면 원시데이터의 정책에 의존하게 되므로 바람직하지 않음
- 획득이 용이하더라도 서비스 활용 측면에서 데이터를 활용하기 위해 가공 처리에 많은 비용이 드는 데이터는 선정하기 어려움
- 직접 산출이 어려운 경우 획득 난이도 측면에서 트래픽량과 저장 처리 장치의 용량 등을 고려 대상,획득 대상의 대안 필요

<br>

### 5.2.2 데이터 균형과 다양성

- 개체의 다양성
- 목적 및 상황의 다양성
- 시간별, 종류별, 사람별, 지역별 다양성

<br>

### 5.2.3 신뢰성

- 데이터의 품질이 신뢰할 수 있는지 검토

<br>

### 5.2.4 법 제도 준수

- 개인정보 및 사생활 보호가 필요한 항목 획득 시, 개인정보보호법 등에 따라 적절한 법적, 기술적 절차를 거친 데이터를 활용하며, 그렇지 않은 데이터는 정제 과정에서 처리될 수 있도록 함

## 5.3 원시 데이터 전처리

### 5.3.1 전처리 단계

#### 5.3.1.1 추출 대상 확인

- 메타 정보
- 주석 대상 텍스트
  - 주석 대상 표현 포함 텍스트 여부
  - 텍스트 길이 범위

<br>

#### 5.3.1.2 정제 대상 확인

- 숫자, 외국어, 기호, 이모지
- 띄어쓰기, 맞춤법, 오탈자
- 개인 정보
- **문장 분리**

<br>

#### 5.3.1.3 불필요 요소 제거 및 변환

- **개인정보 비식별화**
- 비윤리적 표현 정제

## 5.4 원시 데이터 가공

### 5.4.1 주석 (annotation, labeling)

- 원시 데이터를 가공하여 원하는 정보를 부착하는 작업
- 텍스트를 단순히 분류하여(긍부정, 주제) 해당 분류를 텍스트에 삽입하거나 개체명, 관계 정보 등의 정보를 문자열에 직접 주석할 수 있음

<br>

- 주석 시에는 다양한 도구(tool)가 사용됨

<br>

### 5.4.2 주석 도구의 종류

#### 5.4.2.1 구글 스프레드 시트

- 여러 명의 작업자 동시 작업 가능
- 작업과 동시에 저장
- 데이터 관리 용이
- csv 형식으로 export가능

<br>

#### 5.4.2.2 구글 폼

- 단순 분류 문제 등 복잡한 주석 도구가 필요하지 않은 경우에 적합
- 결과를 구글 스프레드 시트로 확인할 수 있음
- 작업자 모집에도 활용
