## 2.2 데이터 관련 용어 정리

### 2.2.1 텍스트 (text)

- 주석, 번역, 서문 및 부록 따위에 대한 본문이나 원문
- 『언어』
  - 문장보다 더 큰 문법 단위
- 문장이 모여서 이루어진 한 덩어리의 글을 이른다.

<br>

### 2.2.2 말뭉치 (corpus, plural corpora)

- corpus - 단수형, corpura - 복수형
- 말뭉치 (이상섭, 1988)
  - 어떤 기준으로든 한 덩어리로 볼 수 있는 말의 뭉치 (한 저작자의 저작 전부, 특정 분야 저작 전체)
- **텍스트 아카이브 (text archive) vs 말뭉치 (corpus)**
  - 텍스트 아카이브 (text archive)
    - 텍스트를 단순히 모아놓고 저장한 것
    - 특별한 기준이나 조건이 없다.
  - 말뭉치 (corpus)
    - 특별한 기준에 맞춰 선별(selectec), 구조화(structured), 설계(designed)된 것을 의미한다.
    - 하지만 일반적으로 언어 데이터를 말뭉치라고 표현하는 경우가 많다.

<br>

### 2.2.3 데이터 (data)

- 『정보·통신』
  - 컴퓨터가 처리할 수 있는 문자, 숫자, 소리, 그림 따위의 형태로 된 정보
- 말뭉치 데이터 (corpus data)
  - 말뭉치 자체
- 말뭉치의 데이터 (data from corpus)
  - 용례 색인 결과, 연어 추출 결과, 통계 분석 결과

<br>

### 2.2.4 주석

- 주석
  - tag, label, annotation
  - 명사로 사용된 경우
  - 기호로서 원래 텍스트에 대해서 붙여놓은 것
- 주석하다
  - tagging, labeling, annotating
  - 행위
- ~~분석하다: anaylsis~~

<br>

- **형태소 분석기** vs 형태소 주석기
  - 영어는 형태소 분석기를 POS(Part of speech) **tagger** (analyzer라고 하지 않음)
  - 분석기라는 용어가 사용된 이유? -> 언어 사실을 분석하고자 하는 열망 반영

<br>

- 형태소 분석기의 기능
  - Segmentation: 형태소 단위로 문장을 분할
  - Tagging: 분할된 문자열에 대해 형태소 태그 부여

<br>

### 2.2.5 언어학의 연구 분야

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<img src='https://drive.google.com/uc?id=1wd6LAApHz5ew_77tsJSO1cKKXEeKUUAd' width=800/>

- 언어의 기본 단위
  - 음성, 음운, 음절, 언어, 단어(어휘), 구, 절, 문장, 텍스트, 말뭉치
- 음성과 음운, 음절의 경우 원래 자연어처리 task에서 다루는 분야는 아니다.
  - 음성학이나 음운론에서 다루는 단위
- 어절은 약간 애매한 위치에 있다.
- 자연어처리 관점에서 관심을 가져야 할 연구 분야들은 다음과 같다.
  - 형태론, 통사론, 텍스트 언어학, 말뭉치 언어학, 의미론, 화용론

<br>

### 2.2.6 텍스트 데이터의 기본 단위

- 영어 말뭉치의 계량 단위
  - 단어(=띄어쓰기 단위), 문장, 발화
- 한국어 말뭉치의 계량 단위
  - **어절**(=띄어쓰기 단위), 문장, 발화
  - 어절 단위를 사용하는 것이 대부분이다.

<br>

- 영어의 단어와 한국어의 어절은 둘 다 띄어쓰기 단위를 기본으로 한다는 공통점이 있다.
- 하지만 한국어는 단어가 곧 띄어쓰기가 아니라는 점이 영어와의 차이점이다.

<br>

- 한국어의 "단어"는 9품사로 분석된다.
  - 명사, 수사, 대명사, 동사, 형용사, 관형사, 부사, 조사, 감탄사
- 이 중 "조사"는 체언(명사, 수사, 대명사)와 꼭 붙어서 사용되기 때문에 띄어쓰기 단위와 단어의 단위가 일치하지 않는다.
- 또한 "어미"는 하나의 품사로 인정되지 않으며 형태 단위이므로 독립된 단어가 아니다.
  - 어미는 단어가 아니지만 조사는 단어이다.

<br>

- 품사
  - 단어를 **문법적 성질의 공통성**에 따라 몇 갈래로 묶어 놓은 것
- 품사 분류의 기준
  - 의미 (뜻, meaning)
  - 기능 (구실, function)
  - 형식 (꼴, form)
- 품사 중에서 동사와 형용사를 제외하고는 형태가 유지된다.
  - 동사와 형용사는 활용을 하면서 형태가 변한다.

<br>

### 2.2.7 타입 (type) & 토큰 (token)

- 계량의 단위

<br>

- 자연어 전처리의 일반적인 과정
  - 토큰화 (tokenization)
    - 토큰 단위로 텍스트를 잘라주는 것 (segmentation)
  - 표제어 추출 (lemmatization) or 품사 주석 (POS(part of speech) tagging)
    - 표제어: 사전에 등재되어 있는 말 (기본 형태)
    - 표제어 추출 시 조사는 대상에서 제외된다.
- 토큰 (token)
  - 언어를 다루는 가장 작은 기본 단위
  - 단어(word), 형태소(morpheme), 서브워드(subword)
- 타입 (type)
  - 토큰의 대표 형태

<br>

- 한국어 예시: "이 사람은 내가 알던 사람이 아니다."
  - 토큰화
    - **이 사람** 은 내 가 알 더 ㄴ **사람 이** 아니 다`
  - 표제어 추출 (Text Mining에서 사용됨)
    - 이, 사람, 나("내"의 원형) 알다, 아니다
  - 품사 주석 (형태소 분석)
    - 이/MM 사람/NNG+은/JX 나/NP+가/JKS 알/VV+더/EP+ㄴ/ETM 사람/NNG+이/JKS 아니/VA+다/EF
  - 토큰 수: 12개
  - 타입 수: 10개

<br>

- 영어 예시: "**She** is gone but **she** used to be mine."
  - 토큰화
    - **She** is gone but **she** used to be mine
  - 표제어 추출
    - She, be, go, but, use, to, mine
  - 품사 주석
    - she_PRP is_VBZ gone_VBN but_IN she_PRP used_VBD to_TO be_VB mine_JJ
  - 토큰 수: 9개
  - 타입 수: 8개

<br>

### 2.2.8 N-gram

- 연속된 N개의 단위
- 단위: 글자, 형태소, 단어, 어절 등으로 사용자가 지정할 수
있음

<br>

- 글자수 bi-gram
  - 흔들리는 꽃들 속에서 네 샴푸향이 느껴진거야
  - 흔+들, 들+리, 리+는, 는+꽃, 꽃+들 ...
- 형태소 bi-gram
  - 흔들리는 꽃들 속에서 네 샴푸향이 느껴진거야
  - 흔들리+는, 는+꽃, 꽃+들, 들+속, 속+에서 ...
- 어절 bi-gram
  - 흔들리는 꽃들 속에서 네 샴푸향이 느껴진거야
  - 흔들리는+꽃들, 꽃들+속에서, 속에서+네 ...

<br>

### 2.2.9 표상 (representation)

- 대표로 삼을 만큼 상징적인 것.
- `표상-하다 「001」「동사」 【…을】`
  - 추상적이거나 드러나지 아니한 것을 구체적인 형상으로 드러내어 나타내다.
- 자연어처리 분야에서 표현으로 번역하기도 하나, 자연어를 컴퓨터가 이해할 수 있는 기법으로 표시한다는 차원에서 표상이 더 적합
- 표시를 통해 재현 과정을 통해 나타내는 작업
- 사전학습모델(PLM, pre-trained language model), word2vec 등

## 2.3 자연어처리 데이터 형식

### 2.3.1 HTML (Hypertext Markup Language)

- 우리가 보는 웹페이지가 어떻게 구조화되어 있는지 브라우저로 하여금 알 수 있도록 하는 마크업 언어
- 보통 웹페이지를 크롤링한 자료는 HTML 형식으로 되어 있음
- 파싱 라이브러리(beautifulsoup 등)을 통해 태그를 제외한 순수한 텍스트만 추출하여 사용

<br>

### 2.3.2 XML (Extensible Markup Language)

- 사람과 기계가 동시에 읽기 편한 구조
- 다른 특수한 목적을 갖는 마크업 언어를 만드는데 사용하도록 권장하는 다목적 마크업 언어
- `<>` 안에 태그 정보를 부여
  - 여는 태그(`<*>`)와 닫는 태그(`</*>`)로 구성
- **HTML과의 차이**
  - **HTML은 태그가 지정되어 있으나 XML은 사용자가 임의로 지정하여 사용할 수 있음**

<br>

### 2.3.3 JSON과 JSONL

- JSON: JavaScript Object Notation
- JSONL: JavaScript Object Notation Lines

<br>

- 속성-값 쌍(attribute­value pairs and array data type(or any other serializable value)) 또는 "키-값 쌍"으로 이루어진 데이터 오브젝트를 전달하기 위해 인간이 읽을 수 있는 텍스트를 사용하는 개방형 표준 포맷
- 기본 자료형
  - 수, 문자열, 불린, 배열, 객체, null
- JSONL은 JSON을 한 줄(Line)으로 반든 것

<br>

### 2.3.4 CSV와 TSV

- CSV (comma-separated values)
  - 몇 가지 필드를 쉼표(`,`)로 구분한 텍스트 데이터 및 텍스트 파일
- TSV(tab-separated values)
  - 몇 가지 필드를 탭(`\t`)로 구분한 텍스트 데이터 및 텍스트 파일
- 구분자(dlimiter)의 차이
