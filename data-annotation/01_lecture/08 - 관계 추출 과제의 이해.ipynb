# 8. 관계 추출 과제의 이해

## 강의 소개

- 관계 추출 과제에 대해 이해합니다.
    - 개체명 인식과 개체명 연결이 어떻게 관계 추출 태스크에 활용되는지 이해합니다.
    - KLUE에서 활용된 CoNLL 2003의 태그 체계가 어떻게 구성되어 있는지 파악합니다.

### 8.1.2 개체명 인식 (NER, Named Entity Recognition)

- 개체명 (Entity)
  - 인명, 지명, 기관명 등과 같은 <mark>고유명사</mark>나 <mark>명사구</mark>를 의미
- cf) 고유명사
  - 일반명사와는 다른 개념으로 보편적으로 사용되지 않고 특정 대상만 지칭하는 명사
  - ex) 스탬플러 - 일반명사 vs 호치키스 - 고유명사
  - ex) 콜라 - 일반명사 vs 코카콜라 - 고유명사
- cf) 구 (phrase)
  - 하나 이상의 단어가 모여서 만들어진 언어 단위
  - 하나의 주어, 목적어, 서술어구로 사용될 수 있음
  - 온전한 "절"을 이룰 수 없는 상태일 때를 "구"라고 한다.
- 개체명 인식 태스크
  - 문장을 분석 대상으로 삼아서 문장에 출현한 개체명의 경계를 인식하고, 각 개체명에 해당하는 태그를 주석함

<br>

- KLUE에서는 국제적인 기준에서 가장 널리 알려진 <mark>CoNLL 2003</mark>의 태그 체계 및 Stanford NER을 바탕으로 국내 TTA표준 지침의 주석 가이드라인에 따라 데이터를 구축함.
  - PS - 사람
  - LC - 지역
  - OG - 기관
  - DT - 날짜
  - TI - 시간
  - QT - 수량

<br>

### 8.1.3 관계 추출 (RE, Relation Extract)

- 관계 추출 (Relation Extract)
  - 문장에서 나타난 개체명 쌍(Entity Pair)의 관계(Relation)을 판별하는 태스크
  - Relation Extract를 Slot Filling이라고 부르기도 한다.
- 개체명 쌍은 관계의 주체(Subject)와 대상(Object)로 구성됨

<br>

- KLUE에서는 <mark>TACLED</mark>에 기반하여 30개 관계 Class를 설정하여 데이터를 구축함.

<br>

### 8.1.4 개체명 연결 (EL, Entity Linking)

- 개체명을 인식(Named Entity Recognition)하고 모호성을 해소(Named Entity Disambiguation)하는 과제를 결합한 것
  - NER + NED
- 텍스트에서 추출된 개체명을 <mark>지식 베이스(knowledge base)</mark>와 연결하여 모호성을 해소함
- 관련 데이터셋
  - AIDA CoNLL-YAGO Dataset
  - TAC KBP English Entity Linking Comprehensive and Evaluation Data 2010

<br>

- cf) Knowledge base
  - 위키피디아 또는 자체적으로 제작해놓은 지식 베이스를 의미

<br>

- 특정 단어가 여러 개의 개체명을 가질 수 있을 것이다.
  - ex) "Kashmir": song, region
- 여러 개의 개체명 중 어떤 것이 맞는 지 확실히 하기 위해 Knowledge base를 확인하여 연결을 해준다.

## 8.4 데이터 구축 시 문제점

### 8.4.1 NER

#### 8.4.1.1 2개 이상의 태그로 주석될 수 있는 개체명

- 맥락에 기반한 주석 처리
- 작업자마다 일치도가 떨어지는 문제가 발생할 수 있음
- ex) "<mark>서울시</mark>는 정책을 발표했다." vs "그 카페는 <mark>서울시</mark> 서대문구 연희동에 있다."


<br>

#### 8.4.1.2 주석 대상의 범주

- 구체적 범주 및 기준 명시
- 제대로 명시하지 않은 경우 작업자마다 결과물이 달라질 수 있다.
- ex) A급, B급, C급, 삼류(3류)

<br>

### 8.4.2 RE

#### 8.4.2.1 한국어 데이터 현실에 맞지 않는 주석

- 태그 통폐합 및 추가
- ex) 지역 관련 태그 통합, 사람, 기관의 작품 및 생산물 관련 태그 추가


<br>

#### 8.4.2.2 KB(Knowledge base)의 활용에 대한 고민

- KB를 일부만 활용하고 나머지는 작업자들의 판단에 맡김

<br>

### 8.4.3 EL

#### 8.4.3.1 적합한 KB(Knowledge base) 선정의 문제

- 현재 AI HUB에 공개된 KB의 경우 제한적인 저작권 아래서 활용이 가능함
- 위키 데이터를 활용하여 자체적인 지식베이스를 구축하여 활용하거나, 서비스 도메인에 맞는 지식베이스를 구축하여 활용할 수 있음
- 지식베이스를 구축하는 것 자체가 많은 비용과 자원이 드는 일이므로 이에 대한 대비가 필요함

<br>

## 8.5 이러한 데이터를 만드는 이유

- NER, RE, EL은 기본적으로 비구조화된 텍스트에서 정보를 추출하여 구조화하려는 것이 목적
- 따라서 이 과정에서 지식 베이스가 활용되기도 하고, 이 결과물이 지식 베이스가 되기도 함
- 정보처리의 관점에서 구조화된 정보의 활용도가 높기 때문에 이러한 시도는 앞으로도 계속 될 것

<br>
