## 1.6 lambda, map, apply

### 1.6.1 Lambda 함수

- 한 줄로 함수를 표현하는 익명 함수 기법

```
lambda argument : expression
```

```python
lambda x, y: x + y
```

````python
# lambda 사용 x
def f(x, y):
    return x + y
f(1,4)
````

````python
# lambda 사용 o
f = lambda x,y: x+y
f(1,4)
````

````python
# 하나의 argument 만 처리하는 lambda 함수 (1)
f = lambda x: x/2
f(3)
````

````python
# 하나의 argument 만 처리하는 lambda 함수 (2)
f = lambda x: x**2
f(3)
````

````python
# 이름을 할당하지 않는 lambda 함수
(lambda x: x+1)(5)
````

<br>

### 1.6.2 map 함수 (자주 사용!)

- 함수와 sequence형 데이터를 인자로 받아 각 element 마다 입력받은 함수를 적용하여 **list로 반환**
- 일반적인 함수를 lambda 형태로 표현함

```
map(function, sequence)
```

````python
ex = [1, 2, 3, 4, 5]
f = lambda x: x**2
list(map(f, ex))
````

````python
# 두 개 이상의 argument 가 있을 때는 두 개의 sequence형을 써야 함
ex = [1, 2, 3, 4, 5]
f = lambda x, y: x + y
list(map(f, ex, ex))
````

````python
# 익명 함수 그대로 사용할 수 있음
ex = [1, 2, 3, 4, 5]
list(map(lambda x: x+x, ex))
````

<br>

#### 1.6.2.1 map for series

- pandas 의 series type 의 데이터에도 map 함수 사용 가능
- function 대신 dict, sequence형 자료 등으로 대체 가능

````python
s1 = Series(np.arange(10))
s1
````

````python
s1.map(lambda x: x**2)
````

````python
# dict type으로 데이터 교체 (없는 값은 NaN) 많이 사용!
z = {1: 'A', 2: 'B', 3: 'C'}
s1.map(z)
````

````python
# 같은 위치의 데이터를 s2로 전환
s2 = Series(np.arange(10, 20))
s1.map(s2)
````

### 1.6.3 apply 함수

#### 1.6.3.1 apply for dataframe

- map과 달리, dataframe 에 있는 모든 series(column)에 해당 함수를 적용
- 입력값이 series 데이터로 입력받아 handling 가능

<br>

- 내장 연산 함수를 사용할 때도 똑같은 효과를 거둘 수 있음

### 1.8.3 grouped

#### 1.8.3.2 grouped apply

- 추출된 group 정보에는 세 가지 유형의 apply 가 가능
- Aggregation
  - `agg()`
  - 요약된 통계정보를 추출
- Tranformation
  - `transform()`
  - 해당 정보를 변환
- Filtration
  - `filter()`
  - 특정 정보를 제거하여 보여주는 필터링 기능

<br>

##### 1.8.3.2.1 aggregation

````python
grouped.agg(sum)
````

````python
grouped.agg(np.mean)
````

````python
# 특정 컬럼에 여러 개의 function 을 apply 할 수도 있음
grouped['Points'].agg([np.sum, np.mean, np.std])
````

<br>

##### 1.8.3.2.2 tranformation

- aggregation 과 달리 key값 별로 요약된 정보가 아니라 개별 데이터의 변환을 지원

````python
df
````

````python
score = lambda x: (x)
grouped.transform(score)
````

<br>

- 단 `max` 나 `min` 처럼 Series 데이터에 적용되는 데이터들은 key값을 기준으로 grouped 된 데이터 기준

````python
score = lambda x: (x.max()) # 각 split 된 group에서 max 추출
grouped.transform(score)
````

````python
score = lambda x: (x-x.mean()) / x.std() # group 별 정규화
grouped.transform(score)
````

<br>

##### 1.8.3.2.3 filter

- 특정 조건으로 데이터를 검색할 때 사용
- filter 안에는 boolean 조건이 존재해야 함
- 자주 사용됨

````python
df.groupby('Team').filter(lambda x: len(x) >= 3).sort_values('Team') # len(x) : grouped 된 dataframe 갯수
````

````python
df.groupby('Team').filter(lambda x: x['Rank'].sum() > 2).sort_values('Team')
````

````python
df.groupby('Team').filter(lambda x: x['Points'].sum() > 1000).sort_values('Team')
````

````python
df.groupby('Team').filter(lambda x: x['Rank'].mean() > 1).sort_values('Team')
````

## 1.10 Merge & Concat

### 1.10.1 Merge

#### 1.10.1.1 merge basic

- SQL에서 많이 사용하는 merge와 같은 기능
- sql 의 join 기능
- 두 개의 데이터(ex. dataframe)를 하나로 합침

````python
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_a
````

````python
raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b
````

````python
# subject_id를 기준으로 merge
pd.merge(df_a, df_b, on='subject_id')
````

````python
# 두 dataframe이 column 이름이 다를 때
pd.merge(df_a, df_b, left_on='subject_id', right_on='subject_id')
````

<br>

#### 1.10.1.2 Join Method

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<img src='https://drive.google.com/uc?id=1QOArH92-W7aUpamRPGtpr-yZvf27R6ye' width=500/>

````python
# left join
pd.merge(df_a, df_b, on='subject_id', how='left')
````

````python
# right join
pd.merge(df_a, df_b, on='subject_id', how='right')
````

````python
# full(outer) join
pd.merge(df_a, df_b, on='subject_id', how='outer')
````

````python
# inner join
pd.merge(df_a, df_b, on='subject_id', how='inner')
````

<br>

#### 1.10.1.3 Index based join

- `right_index=True` 와 `left_index=True` 지정하여 각각의 index 값을 기준으로 합침

````python
pd.merge(df_a, df_b, right_index=True, left_index=True)
````
