## 1.4 Why PyTorch?

- Define by Run 의 장점
  - 즉시 확인 가능 -> pythonic code (가장 큰 이유)
- GPU support, Good API and community
- 사용하기 편한 장점이 가장 큼
- TF는 production 과 saclability 의 장점

<br>

## 1.5 PyTorch 의 3가지 특징

- Numpy
  - Numpy 구조를 가지는 Tensor 객체로 array 표현
- AutoGrad (자동미분)
  - 자동미분을 지원하여 DL 연산을 지원
- (DL) Function
  - 다양한 형태의 DL을 지원하는 함수와 모델을 지원
  - Dataset
  - Multi-GPU
