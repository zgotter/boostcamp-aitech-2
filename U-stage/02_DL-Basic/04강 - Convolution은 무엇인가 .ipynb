# 4. Convolution 은 무엇인가?

## 4.0 계산식

- 출력의 크기

$$
\text{출력 크기} = \frac{\text{입력 크기} - \text{커널 크기} + 2 \times \text{padding}}{\text{stride}} + 1
$$

- 파라미터 갯수

$$
\text{파라미터 갯수} = \text{커널 길이} \times \text{커널 높이} \times \text{입력 채널 수} \times \text{출력 채널 수}
$$

## 4.1 Convolution

### 4.1.1 Convolution 의 등장

- 두 함수를 잘 섞어주는 operator 로서 정의



- Continuous convolution

$$
(f * g)(t)=\int f(\tau) g(t-\tau) d \tau=\int f(t-\tau) g(t) d \tau
$$

- Discrete convolution

$$
(f * g)(t)=\sum_{i=-\infty}^{\infty} f(i) g(t-i)=\sum_{i=-\infty}^{\infty} f(t-i) g(i)
$$

- 2D image convolution

$$
(I * K)(i, j)=\sum_{m} \sum_{n} I(m, n) K(i-m, j-n)=\sum_{m} \sum_{n} I(i-m, i-n) K(m, n)
$$

- $I$ : 전체 이미지
- $K$ : Convolution filter

## 4.4 Convolutional Neural Networks

- CNN consists of convolution layer, pooling layer, and fully connected layer.
  - Convolution and pooling layers: feature extractioni
  - Fully connected layer: decision making (e.g., classification)

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<img src='https://drive.google.com/uc?id=1NoX3QxtYQWyMPY2D6D64AGnoVsZZLPy-' width=600/>

- 최근에는 fully connected layer 가 없어지거나 최소화시키는 것이 추세이다.
  - 이러한 추세의 원인은 파라미터의 숫자에 dependent 한다.
  - 학습해야 하는 파라미터의 숫자가 늘어날 수록 학습이 어렵고 일반화 성능이 떨어진다.
  - 즉, 아무리 학습을 잘 시켜도 실제 문제에 deploy 하면 좋은 성능을 못 낼 수 있다는 것이다.
  - 그래서 convolution layer 를 deep 하게 쌓지만 파라미터의 갯수는 줄이는 것에 집중한다.
  - 이를 위한 여러 가지 테크닉이 들어가게 된다.

- 어떤 neural network 를 봤을 때 network 의 layer 별로 몇 개의 파라미터로 이루어져 있고 전체 파라미터의 숫자가 몇 개 인지를 파악하고 있는 것이 중요하다.
