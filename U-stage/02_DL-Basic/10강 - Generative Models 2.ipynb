### 10.2.1 Variational inference (VI)

- The goal of VI is to optimize the **variational distribution** that best matches the **posterior distribution**.

<br>

- Posterior distribution
  - 사후 분포
  - $p_{\theta}(z \mid x)$
- Variational distribution
  - 변분 분포
  - $q_{\phi}(z \mid x)$

<br>

- In particular, we want to find the **variational distribution** that minimizes the KL divergence between the true posterior.

### 10.4.3 GAN Objective

- A two player minimax game between **generator** and **discriminator**.
- For **discriminator**

$$
\max _{D} V(G, D)=E_{\mathbf{x} \sim p_{\text {data }}}[\log D(\mathbf{x})]+E_{\mathbf{x} \sim p_{G}}[\log (1-D(\mathbf{x}))]
$$

- where the **optimal discriminator** is

$$
D_{G}^{*}(\mathbf{x})=\frac{p_{\text {data }}(\mathbf{x})}{p_{\text {data }}(\mathbf{x})+p_{G}(\mathbf{x})}
$$

- For **generator**

$$
\min _{G} V(G, D)=E_{\mathbf{x} \sim p_{\text {data }}}[\log D(\mathbf{x})]+E_{\mathbf{x} \sim p_{G}}[\log (1-D(\mathbf{x}))]
$$
