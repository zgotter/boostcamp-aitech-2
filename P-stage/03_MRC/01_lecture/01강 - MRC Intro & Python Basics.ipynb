### 1.1.5 MRC의 평가 방법

#### 1.1.5.1 Exact Match(EM) / F1 Score

- For **extractive** answer and **multiple-choice** answer datasets
- 질문에 대한 답이 지문이 존재하는 경우의 Extractive Answer task를 평가할 때 사용
- 주어진 보기 중에 답변을 고르는 Multiple-choice task를 평가할 때 사용

<br>

- Descriptive/Narrative Answer task에 대한 평가 지표로 사용하기는 힘들다.

<br>

**Exact Match(EM) or Accuracy**

- 예측한 답과 ground-truth이 정확히 일치하는 샘플의 비율
- 정확도(Accuracy)와 똑같다고 볼 수 있다.
  - 일치하면 1, 일치하지 않으면 0

$$
\text{Number of correct samples} \over \text{Number of whole samples}
$$

<br>

**F1 Score**

- 예측한 답과 ground-truth 사이의 token overlap을 soft하게 계산하기 위해 F1을 사용
- 조금 겹치는 부분에 대해 partial score를 부여

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<img src='https://drive.google.com/uc?id=1j98Ug8BIivYuG0_ljfKMPpTnUD7MB1Ky' width=400/>

<br>

#### 1.1.5.2 ROUGE-L / BLEU

- For **descriptive** answer datasets
- Descriptive/Narrative Answer task를 평가할 때 사용
  - F1 score 단어의 overlap만 보기 때문에 한계가 있다.
- Ground-truth과 예측한 답 사이의 overlap을 계산

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<img src='https://drive.google.com/uc?id=1jF5YcBVaUfM9jEsXOsb37GboCK3QC2DB' width=400/>

<br>

**ROUGE-L Score**

- 예측한 값과 ground-truth 사이의 overlap recall
  - 단어 사이의 overlap 뿐만 아니라 n-gram token의 overlap도 평가한다.
- ROUGE-<U>L</U> -> LCS (Longest common subsequence) 기반
- 가장 긴 common subsequence의 비율을 점수로 사용
  - 정답의 길이가 15라고 하자.
  - 예측값과 정답이 겹치는 subsequence의 길이가 10이다.
  - 이 경우 ROUGE-L는 10/15인 67%가 된다.

<br>

**BLEU (Bilingual Evaluation Understudy)**

- 예측한 답과 ground-truth 사이의 precision
- BLEU-<U>n</U> -> uniform <U>n</U>-gram weight
  - n이라는 숫자를 정의하여 n-gram의 토큰을 평가에 사용

### 1.2.3 UTF-8 (Unicode Transformation Format)

- UTF-8은 현재 가장 많이 쓰는 인코딩 방식
- 문자 타입에 따라 다른 길이의 바이트를 할당한다.
  - 1 byte
    - Standard ASCII
  - 2 bytes
    - Arabic, Hebrew, most European scripts
  - 3 bytes
    - BMP(Basic Multilingual Plane)
    - 대부분의 현대 글자 (한글 포함)
  - 4 bytes
    - All Unicode characters
    - 이모지 등

### 1.3.5 KorQuAD 예시

- Extractive 방식이므로 지문 내에 답변에 해당하는 단어가 포함되어 있다.
- `answer_start`는 정답에 해당하는 단어가 지문 내의 몇 번째 인덱스에 등장하는 지를 나타낸다.
  - 정답 단어가 한 번 등장한다면 문제가 되지 않지만 경우에 따라선 정답 단어가 지문에서 여러 번 등장할 수 있다.
  - 그렇지만 답변을 유추할 수 있었던 문장이 특정 하나의 문장이였을 수 있다.
  - 이러한 부분이 학습할 때 중요하다.
- 이와 같이 `answer_start` 인덱스가 주어진 것을 strong supervision이라고 하고, 그렇지 않은 경우를 distant supervision이라고 한다.

- 실제로 하나의 질문에 대해 답이 여러 개일 수 있기 때문에 `answers`의 value가 리스트 형태로 되어 있다.
