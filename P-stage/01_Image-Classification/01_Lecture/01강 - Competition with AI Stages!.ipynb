# 1. Competition with AI Stages!

> 드디어 P stage를 시작합니다!  
첫 강의에서는 P stage의 의미와 컴퍼티션 소개, 그리고 P stage와 컴퍼티션 진행이 어떤 관련이 있는지 알아보도록 하겠습니다.  
또한, 컴퍼티션의 각 구성요소가 데이터 파이프라인 과정에서 어떠한 의미를 지니는 지를 Deep하게 다뤄보겠습니다.

## 1.1 Goals

- U Stage 에서 경험한 이론을 바탕으로 실제 데이터와 코드베이스를 통한 이해
- Competition 형태의 실습을 통해 점진적인 모델 성능 향상을 경험
- 머신러닝 파이프라인의 한 부분을 경험


<br>

## 1.2 Competition

- "주어진 데이터를 이용해 원하는 결과를 만들기 위한 가장 좋은 방법이란?"
  - 댓글 데이터를 이용한 악성 댓글 분류
  - 폐 CT 사진을 이용한 폐 질병 예측
  - 위성 항공 사진을 이용한 구름 종류 분류
  - ...
