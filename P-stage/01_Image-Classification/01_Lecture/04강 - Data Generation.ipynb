### 4.2.2 모델에 먹이(Data)를 주다

- 혹시 모델 학습을 할 때 아래와 비슷한 오류를 범하고 있진 않는가?
  - 모델이 처리할 수 있는 양이 정해져 있다고 하자 (20 batch/s)
  - generator 속도가 모델의 처리 속도보다 느린 경우
    - 아무리 모델의 처리 속도가 빠르더라도 generator 속도 만큼만 처리 된다.
  - generator 속도가 모델의 처리 속도보다 빠른 경우
    - 아무리 generator의 처리 속도가 빠르더라도 모델의 속도 만큼만 처리 된다.

### 4.2.3 Example

- Dataset 생성 능력 비교

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<img src='https://drive.google.com/uc?id=17EShDSv2ygmGMvxlvdkrpNHIIXPPesN6' width=800/>

- resize 를 추가(300x300 -> 1024x1024)했을 때 처리 시간이 증가한다.
  - 여러 가지 실험을 해보면서 속도를 줄여야 한다.
- resize 를 미리 하고 rotation 을 하면 시간은 더 오래 걸린다.
  - 300x300 을 rotation 하는 것보다 1024x1024 를 rotation 하는 것이 더 힘들 것이다.
- `transforms` 를 적용하는 순서를 정하는 것 조차 튜닝의 대상이 된다.
