### 5.3.5 `nn.Module` Family

- `nn.Module` 을 상속받은 모든 클래스의 공통된 특징

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
<img src='https://drive.google.com/uc?id=16EsK3Oen_YXC_ZYDHpZsENDdIPnWW15y' width=500/>

- 모든 `nn.Module` 은 child modules 를 가질 수 있다.
  - 내 모델을 정의하는 순간, 그 모델에 연결된 모든 module 을 확인할 수 있다.
- 모든 `nn.Module` 은 `forward()` 함수를 가진다.
  - 내가 정의한 모델의 `forward()` 를 한 번만 실행한 것으로 그 모델의 forward에 정의된 모듈 각각의 `forward()` 가 실행된다.
