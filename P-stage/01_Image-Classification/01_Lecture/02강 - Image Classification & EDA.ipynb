# 2. Image Classification & EDA

> 대부분의 태스크와 P stage 과정에서 빼놓을 수 없는 EDA에 대해서 이야기 합니다.  
EDA의 의미와 EDA를 어떻게 해야 하는지를 다뤄보려고 합니다. (특정한 EDA 방법은 따로 설명하지 않습니다.)

> 그 다음, Image Classification의 과정을 아주 Rough하게 설명합니다. 우리는 문제 정의를 할 때 Input과 Output을 명확히 정의할 필요가 있습니다. 어떻게 보면 당연하다고 느껴질 수 도 있지만, 아주 Rough하게 우리가 하려는 태스크의 시스템을 명확히 설정하는 것이 중요합니다.  

> Image(Input) -> Classification Model  -> Class(Output)

> 그리고, 베이스라인 활용법에 대해서 설명합니다.
